```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def check_channels(image_path):
    # Read the image
    img = mpimg.imread(image_path)
    
    # Display the image
    plt.imshow(img)
    plt.axis('off')
    plt.title('Image with {} channels'.format(img.shape[2]))
    plt.show()

# Example usage
image_path = "datasets/MNIST/trainingSet/trainingSet/1/img_0.jpg"
check_channels(image_path)
```
